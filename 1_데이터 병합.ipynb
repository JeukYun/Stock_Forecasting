About Dataset <br>
Title: Stock Prices of 500 Biggest Companies by Market Cap (Last 5 Years)

- Date: The date corresponding to the recorded stock market data.
- Open: The opening price of the stock on a given date.
- High: The highest price of the stock reached during the trading day.
- Low: The lowest price of the stock observed during the trading day.
- Close: The closing price of the stock on a specific date.
- Volume: The volume of shares traded on the given date.
- Dividends: Any dividend payments made by the company on that date (if applicable).
- Stock Splits: Information regarding any stock splits occurring on that date.
- Company: Ticker symbol or identifier representing the respective company.

# 원본 데이터, 경제지표 매핑

```python
data.info()
```

```python
# 필요없는 Date 뒤 시간 제거
data['Date'] = data['Date'].str[:10]
data
```

- 금리(DFF) 매핑

```python
data = pd.merge(data, FFE, left_on = 'Date', right_on = 'DATE').drop(columns = 'DATE')
data
```

- 취업지수 매핑
    - 취업지수 : 매주 토요일 발표
    - 주식데이터 : 매주 토,일요일 x

```python
data['year_week'] = data['Date'].astype('datetime64').dt.year.astype('str') + '-' + data['Date'].astype('datetime64').dt.weekofyear.astype('str')
data
```

```python
NPP['year_week'] = NPP['DATE'].astype('datetime64').dt.year.astype('str') + '-' + NPP['DATE'].astype('datetime64').dt.weekofyear.astype('str')
```

```python
NPP # 11월 24일에 발표 된 고용지수는 47주차가 아닌 48주차에 영향을 미치므로 2018-48로 매핑되어야 할 필요...
```

```python
import datetime

NPP['DATE'] = NPP['DATE'].astype('datetime64') + datetime.timedelta(days = 2)
NPP['year_week'] = NPP['DATE'].astype('datetime64').dt.year.astype('str') + '-' + NPP['DATE'].astype('datetime64').dt.weekofyear.astype('str')
```

```python
NPP
```

```python
data = pd.merge(data, NPP, on = 'year_week', how='left').drop(columns = ['year_week', 'DATE'])
data
```

```python
data.info()
```

- 분기별 GDP 매핑

```python
RG['quart'] = RG['DATE'].astype('datetime64').dt.year.astype('str') + '-' + RG['DATE'].astype('datetime64').dt.quarter.astype('str')
RG
```

```python
data['quart'] = data['Date'].astype('datetime64').dt.year.astype('str') + '-' + data['Date'].astype('datetime64').dt.quarter.astype('str')
data
```

```python
data = pd.merge(data, RG, on = 'quart', how = 'left').drop(columns = ['DATE', 'quart'])
data
```

- CPI, 실업률 매핑
    - 두 데이터 모두 월별 데이터므로 한번에 진행

```python
CPI['Ymonth'] = CPI['DATE'].astype('datetime64').dt.year.astype('str') + '-' + CPI['DATE'].astype('datetime64').dt.month.astype('str')
UNR['Ymonth'] = UNR['DATE'].astype('datetime64').dt.year.astype('str') + '-' + UNR['DATE'].astype('datetime64').dt.month.astype('str')
data['Ymonth'] = data['Date'].astype('datetime64').dt.year.astype('str') + '-' + data['Date'].astype('datetime64').dt.month.astype('str')
display(CPI, UNR, data)
```

```python
data = pd.merge(data, CPI, on = 'Ymonth', how = 'left')
data = pd.merge(data, UNR, on = 'Ymonth', how = 'left')
data
```

```python
data = data.drop(columns = ['Ymonth', 'DATE_x', 'DATE_y'])
data
```
