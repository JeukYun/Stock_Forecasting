# 다중공선성 확인

```python
new_AAPL = AAPL[['Open', 'Volume', 'Dividends', 'Stock Splits', 'IR', 'PP', 'GDP', 'CPI', 'UNEMP']]
```

```python
import statsmodels.api as sm 

target = new_AAPL[['Open']]
x_data = new_AAPL.drop(columns = 'Open')
```

```python
# 상수항 추가
x_data1 = sm.add_constant(x_data, has_constant = 'add')
x_data1
```

상수항 추가하는 이유<br>
독립변수가 0일때 종속변수가 0이 나오지는 않음<br>
ex: GDP, CPI..등 여러 변수가 0이라도 주식값은 0일 수 없으므로 <br>
- 모델을 조금 더 유연하게 만들어 줌

```python
# 모든 변수 추가하여 OLS검사 진행
model = sm.OLS(target, x_data1)
fit_model = model.fit()
```

OLS모델 해석 : https://ysyblog.tistory.com/119

```python
fit_model.summary()
```

- No. Observations : 총 표본 수
- R-squared : 결정계수
- F-statistic : F통계량, 도출된 회귀식이 적절한지 판단
    - 0에 가까울수록 적절
- Prob (F-statistic) : 회귀식의 유의미성 판단
    - 0.05이하인 경우 변수끼리 매우 관련이 있음

- coef : 회귀계수
- std err : 계수 추정치의 표준오차(작을수록 좋음)
- t : t-test, 독립변수와 종속변수 사이 상관관계
    - 절대값이 클수록 상관도가 큼
- P>|t| : 독립변수 유의확률
    - p_value가 0.05보다 작아야 유의미한 변수

- Omnibus : 정규성 테스트, 값이 클수록 정규분포를 따름
    - Prob(Omnibus) : 위 검정의 유의성 판단 (0.05이하시 유의)
- Skew, Kurtosis : 왜도(0일수록 대칭)와 첨도(3일수록 정규분포)
- Durbin-Watson : 잔차의 독립성 판단 (1.5~2.5사이일때 잔차는 독립)
    - 0 or 4에 가까울수록 잔차들은 자기상관 존재
- Jarque-Bera (JB) : 정규성검정, 값이 클수록 정규분포 데이터 사용
    - Prob(JB) : 위 검정의 유의성 판단 (0.05 이하시 유의)
- Cond. No. : 다중공선성 검정, 10 이상의 경우 다중공선성이 있다고 판단


[2] The condition number is large, 2.38e+10. This might indicate that there are
strong multicollinearity or other numerical problems. <br>
변수들 간의 높은 상관관계로 인한 다중공선성 우려

# 단계적 선택법을 통한 변수선택

```python
def step_feature_selection(x_data, target):
    feature = x_data.columns.to_list()
    y = target
    selected_feature = []
    sl_enter = 0.01
    sl_remove = 0.01
    sv_per_step = []
    adj_r_square = []
    steps = []
    step = 0

    while len(feature) > 0:
        # remainder : feature에서 selected_feature를 뺀값 (차집합)
        remainder = list(set(feature) - set(selected_feature)) 
        pval = pd.Series(dtype = 'float64', index = remainder)

        # 선택feature를 제외 한 나머지 feature의 최소 pval값 저장 
        for col in remainder:
            x = x_data[selected_feature + [col]]
            x = sm.add_constant(x)
            model = sm.OLS(y,x).fit(disp = 0)
            pval[col] = model.pvalues[col]
        min_pval = pval.min()

        # 최소 pval값이 선정한 0.01보다 작은경우 해당 feature명(index)을 선택
        if min_pval < sl_enter:
            selected_feature.append(pval.idxmin())

            # 반복문 생성, 선정된 feature들의 pval값 확인
            while len(selected_feature) > 0:
                selected_x = x_data[selected_feature]
                selected_x = sm.add_constant(selected_x)
                selected_pval = sm.OLS(y, selected_x).fit(disp = 0).pvalues[1:] # 상수항을 제외 한 pval값 반환
                max_pval = selected_pval.max() # 가장 높은 pval값을 max_pval에 대입 

                # 가장 높은 pval의 값이 선정한 유의수준보다 높은경우 삭제
                # 가장 높은 pval의 값이 유의수준 이하인 경우 반복문 break
                if max_pval >= sl_remove:
                    remove_variable = selected_pval.idxmax()
                    selected_feature.remove(remove_variable)
                else:
                    break
                
            step = step + 1
            steps.append(step)
            r2 = sm.OLS(y,sm.add_constant(x_data[selected_feature])).fit(disp=0).rsquared_adj
            adj_r_square.append(r2)
            sv_per_step.append(selected_feature.copy())

        else:
            break
    return selected_feature
```

```python
selected_variables = step_feature_selection(x_data, target)
print(selected_variables)

# ['CPI', 'USTS', 'DI', 'PP', 'IR', 'OP', 'CLF', 'EPU', 'GDP', 'UNEMP'] 선정
```
